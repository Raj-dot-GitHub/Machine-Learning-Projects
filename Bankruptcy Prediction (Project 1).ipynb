```python
# Oversampling the Target Feature. 
```

```python
import imblearn
```

```python
print(imblearn.__version__)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
X = bankrupt.drop("Bankrupt?", axis = 1)
y = bankrupt['Bankrupt?']
```

```python
oversample = SMOTE()
```

```python
X,y = oversample.fit_resample(X,y)
```

```python
sns.countplot(x=y)
```

```python
# Now our Target Feature is fully balanced.
```

```python
# We have a total of 95 columns as our independent features.
```

```python
# From this 95 columns there are many columns that are not important for our prediction. So we will select best 35 features for building the model.
```

```python
from sklearn.feature_selection import SelectFromModel, mutual_info_classif, f_classif, SelectKBest
```

```python
feature_selection = SelectKBest(f_classif,k=35).fit(X,y)
```

```python
selected_features=X.columns[feature_selection.get_support()]
```

```python
selected_features   # These are the 35 best independent features.
```

```python
# Logistic Regression for Classification
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression(max_iter=100000)
```

```python
log_model.fit(X_train,y_train)
```

```python
predictions = log_model.predict(X_test)
```

```python
# We will check the performance of our logistic model using metrics
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
log_model.score(X_test,y_test)   
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
# So the accuracy of our Logistic Regression model for this problem is 91%.
```

```python
# K Nearest Neighbours.
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
# Here the KNeighborsClassifier takes a parameter 'n_neighbors' meaning number of neighbors to refer.
```

```python
# 'n_neighbors' parameter has a greater influence on our model accuracy.
```

```python
# Here we will try to find the best value for 'n_neighbors' .
```

```python
error_rate = []

for i in range(1,40):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
# So the above code will run a loop which check the error rate for each value of 'K' between 1 & 40 and append the error's in the error_rate list.
```

Now let's see visually how the error rate changes for different values of 'K'.
- Note:- We will select a 'K' value with the least error.

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
# From the above plot we can say that the minimum error is obtained at K = 2.
So using this information we will train our K Neighbor's model.
```

```python
knn = KNeighborsClassifier(n_neighbors=2)

knn.fit(X_train,y_train)
pred = knn.predict(X_test)

print('WITH K=2')
print('\n')
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```

```python
knn.score(X_test,y_test)
```

```python
# So the accuracy of our K-Nearest Neighbors model is 97.39%
```
